```python
# Dependencies and Setup
import pandas as pd
import os
# File to Load (Remember to Change These)
file_path = os.path.join("Resources", "HeroesOfPymoli_Purchase_Data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_path)
purchase_data
```

```python
#Extract the columns for SN and Age
age=pd.DataFrame(purchase_data, columns = ['SN','Age'])

#Extract Unique List
age_unique=age.drop_duplicates()
age_unique
```

```python
# Create the bins in which Data will be held

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
age_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_unique["age group"] = pd.cut(age_unique["Age"], bins, labels=age_labels)
age_unique

age_unique.dtypes
```
